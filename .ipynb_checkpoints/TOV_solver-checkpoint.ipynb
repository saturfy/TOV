```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
#----------------------------------Implementations of classes and functions-----------------------------------------

class EoS:
    """ 
    The constructor read the EoS file from the fname given
    The standard EoS file is 3 rows [nuclear density, energy density, pressure]
        
    interp:_order: the order of the interpolation which is used to calculate e(p) and p(e) from data
    (e(p) and p(e) are only real inverses of each other if the interpolation is 1st order)
        
    units: refers to the units in which the EoS is given, definitions are below
    units=MeV
    [nuclear density[fm^(-3)], energy density [MeV^(4)], pressure [MeV^(4)]]
    units=km
    [nuclear density[fm^(-3)], energy density [km^(-2)], pressure [km^(-2)]]
    """
    unitlist = {
        'MeV':'$\mathrm{MeV}^4$',
        'km' : '$\mathrm{km}^{-2}$'
    
    }
    
    def __init__(self,fname,units,interp_order):
        self.name = fname
        self.units = units
        self.rawdata=np.loadtxt(fname)
        self.p = self.rawdata[::,2]
        self.e = self.rawdata[::,1]
        self.Interpolation(interp_order)
    def VisualizeEoS(self):
        """
        Plots the EoS
        """
        fig = plt.figure()
        eos = fig.add_subplot(1,1,1)
        eos.set_xlabel("p ["+self.unitlist[self.units]+"]")
        eos.set_ylabel("$\epsilon$ ["+self.unitlist[self.units]+"]")
        eos.plot(self.p,self.e,label=self.name)
        plt.legend()
        plt.show(fig)
    def UnitChange(self,convstring):
        """
        Changes the units of the EoS
        convstring is a string: starting units +to+ final units
        """
        f = 197.327 #197 Mev * fm = 1
        e0conv = 0.00132
               
        if (convstring == 'MeVtokm'):
            self.p = f**(-3)*0.001*e0conv*self.p
            self.e = f**(-3)*0.001*e0conv*self.e
            self.units = "km"
        if (convstring == 'kmtoMeV'):
            self.p = f**(3)*1000*e0conv**(-1)*self.p
            self.e = f**(3)*1000*e0conv**(-1)*self.e
            self.units = "MeV"
        
        
    def Interpolation(self,kflag):
        """
        Interpolates the pressure and energy density data to e(p) and p(e) function
        """
        if kflag == 1:
            self.ep = interp1d(self.p,self.e,'linear')
            self.pe = interp1d(self.e,self.p,'linear')
        if kflag == 2:
            self.ep = interp1d(self.p,self.e,'quadratic')
            self.pe = interp1d(self.e,self.p,'quadratic')      
        if kflag == 3:
            ep = interp1d(self.p,self.e,'cubic')
            pe = interp1d(self.e,self.p,'cubic')
        
class solution: 
    """
    Stores the solution of the TOV equation corresponding to the MR diagram
    mr, er, pr: are the profiles of the mass, energy, and pressure as funcitons of radius
    e: the list of the central energy densities
    r: the list of the r values where everything is calculated
    M,R: The mass and radius list corresponding to the values in the e list
    """
    def __init__(self,fname,e,r,M,R,mr,er,pr):
        self.name = fname
        self.mr = mr
        self.er = er
        self.pr = pr
        self.r = r
        self.e = e
        self.M = M
        self.R = R
    def plotMRdiagram(self):
        """
        plots the M-R diagram
        """
        #unit change to mass of the sun units, msun = 1.476 km
        MSun = 1.4766
        
        fig = plt.figure()
        eos = fig.add_subplot(1,1,1)
        eos.set_xlabel("Radius [km]")
        eos.set_ylabel("mass [$M_{\odot}$]")
        eos.plot(self.R,self.M / MSun,label=self.name)
        plt.legend()
        plt.show(fig)
    def mr_profile(self,ind):
        MSun = 1.4766
        
        fig = plt.figure()
        eos = fig.add_subplot(1,1,1)
        eos.set_xlabel("Radius [km]")
        eos.set_ylabel("mass [$M_{\odot}$]")
        eos.plot(self.r,self.mr[ind,:] / MSun,label="$\epsilon_{C} =$" + str(self.e[ind]) + "$ km^{-2}$")
        plt.legend()
        plt.show(fig)
    def er_profile(self,ind):
        MSun = 1.4766
        
        fig = plt.figure()
        eos = fig.add_subplot(1,1,1)
        eos.set_xlabel("Radius [km]")
        eos.set_ylabel("$\epsilon [km^{-2}]$")
        eos.plot(self.r,self.er[ind,:],label="$\epsilon_{C} =$" + str(self.e[ind]) + "$ km^{-2}$")
        plt.legend()
        plt.show(fig)
    def pr_profile(self,ind):
        MSun = 1.4766
        
        fig = plt.figure()
        eos = fig.add_subplot(1,1,1)
        eos.set_xlabel("Radius [km]")
        eos.set_ylabel("p [$km^{-2}$]")
        eos.plot(self.r,self.pr[ind,:],label="$\epsilon_{C} =$" + str(self.e[ind]) + "$ km^{-2}$")
        plt.legend()
        plt.show(fig)       
                
```

```python
# ----------------------------------TOV equation solver implementation----------------------------------------------

#Implementation of the Euler-solver

def eqp(r,p,e,m):
    """
    This function is the right hand side of the TOV equation for the pressure
    """
    return (p + e)*(m + 4 * np.pi * pow(r,3) * p) / (r * (r - 2 * m))

def eqm(e,r1,r2):
    """
    This function calculates the mass of the spherical shell between radius r1 < r2
    """
    return e * (4 * np.pi / 3) *( pow(r2,3) - pow(r1,3) )

def int_TOV_Euler(e0,pe,ep,dr,rmax,pfin):
    """
    Integrates the TOV equation using the Euler-method
    e0: central energy density
    pe, ep : EoS interpolating functions
    dr: step size in radius in km
    rmax: max radius in km
    pfin: the pressure value where the integration stops
    
    These are np.arrays
    rlist: the list of the r values where the pressure energy density is calculated
    elist: list of the enery density values at each r
    plist: list of the enery density values at each r
    mlist: list of the full mass at each r
    """
    rlist = np.arange(dr,rmax,dr)
    elist = np.zeros(len(rlist))
    plist = np.zeros(len(rlist))
    mlist = np.zeros(len(rlist))
    
    #initial condiitons
    mlist[0] = eqm(e0,0,dr)
    elist[0] = e0
    plist[0] = pe(e0)
    fini = len(rlist)-2
    
    # we do not go for the last step because than we go out from index range when we check the p[i+1]
    # at the last step. This is a problem only for small mass, because it is possible that the r goes outside of rmax range
    for i in range(len(rlist)-1):
        #pressure step
        plist[i+1] = plist[i] - dr * eqp(rlist[i],plist[i],elist[i],mlist[i])
        if (plist[i+1] < pfin):
            fini = i
            break
        #energy step
        elist[i+1] = ep(plist[i+1])
        #mass step
        mlist[i+1] = mlist[i] + eqm(elist[i+1],rlist[i],rlist[i+1])      
    
    return np.array([rlist[fini],mlist[fini],e0,rlist,mlist,elist,plist])

```

```python
#---------------------------Implementation of the MR diagram calcualtor----------------------------------------------

def MR(eos,elist,dr,rmax,pfin,method):
    """
    This function calcultes the MR diagram using the central energy densities given in elist, using the method given
    After execution gives back a solution object 
    
    !!!! Everything is in km units after the solution !!!
   
    eos is an EoS object instance which contains all the EoS data
    dr: is the step size in km
    rmax: the maximum radius in km
    pfin: prssure value where the integration stops, typically the minimum pressure in the EoS
    method: euler
    """
    #these are the list which contain all solutions
    elen= len(elist)
    rlist = np.arange(dr,rmax,dr)
    rlen = len(rlist)
    maxmlist = np.zeros(elen)
    maxrlist = np.zeros(elen)
    mr = np.zeros([elen,rlen])
    er = np.zeros([elen,rlen])
    pr = np.zeros([elen,rlen])
    
    #the solution itself 
    if (method == 'euler'):
        for i in range(elen):
            reslist = int_TOV_Euler(elist[i],eos.pe,eos.ep,dr,rmax,pfin) 
            maxrlist[i] = reslist[0]
            maxmlist[i] = reslist[1]
            mr[i,:] = reslist[4]
            er[i,:] = reslist[5]
            pr[i,:] = reslist[6]
             
    #creating the solution object
    sol = solution(eos.name,elist,rlist,maxmlist,maxrlist,mr,er,pr)
    
    return(sol)
```

```python
# You can load an eos from file by providin the file name, this goes into the EoS object
eos = EoS("walmod_x3_x4_Lm_asym.txt","MeV",2)
#You can change the units of the loaded eos to km
eos.UnitChange('MeVtokm')
#Calculate the interpoaltion of e(p) and p(e) in the desired units
eos.Interpolation(2)
#you can plot the eos in the desired units
eos.VisualizeEoS()
```

```python
#select the desired central energy list, for the solutions 
# here we take every 10th element of the energy density list from the EoS
elist = eos.e[np.arange(1,len(eos.e),10)]

#the solver gives back a "solution type object, this takes a few seconds if elist is long"
# arguments (the EOS object, )
meg = MR(eos,elist,0.01,20,eos.p[0],'euler')
```

```python
#explore the solution
#plot the MR diagram
meg.plotMRdiagram()
```

```python
#choose an energy density value and plot the m profile for that star
meg.mr_profile(60)
```

```python
#choose an energy density value and plot the energy density profile for that star
meg.er_profile(60)
```
